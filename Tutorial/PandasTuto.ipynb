# Pandas Series


```python
# Create a Series which is like a column in a table
newSeries = [1, 4, 6, 8, 10]
myVar = pd.Series(newSeries)
myVar
```

```python
# Type of Series
type(myVar)
```

```python
# Labels of Series
myVar[2]
```

```python
# Create Labels Index of Series
myVar = pd.Series(newSeries, index=["A", "B", "C", "D", "E"])
myVar
```

```python
# Get Data by index
myVar['C']
```

```python
# Dictionary as Series
dictSeries = {
    "day1": 420,
    "day2": 478,
    "day3": 500,
}
myVar = pd.Series(dictSeries)
myVar
```

```python
# Using Only First two index
myVar = pd.Series(dictSeries, index=['day1', 'day2'])
myVar
```

```python
# Dictionary as Series
dictSeries = {
    "day1": [420, 4578, 75263, 96965],
    "day2": [478, 1452, 7856, 9632],
    "day3": [500, 600, 700, 800],
}
myVar = pd.Series(dictSeries)
myVar
```

#   Another Methods Pandas

```python
# Nan means empty cells data
# Nat mean empty time stamp data cells
df=pd.DataFrame({
    "name":['Alfred','Batman','Catwoman'],
    "toy":[np.NaN,'Batmobile','Bullwhip'],
    "born":[pd.NaT,pd.Timestamp("1940-04-25"),pd.NaT]
})
df
```

```python
# Dropna returns those rows which rows all cells are not null
df.dropna()
```

```python
# Dropna returns those rows which rows all cells are not null
df.dropna(how='any')
```

```python
# Set all columns Null
df['toy']=np.NaN
df
```

```python
# Drop Coulmn wise values
df.dropna(how='all',axis=1)
```

```python
# Fillna Replace All Empty Values
df['toy'].fillna("Car",inplace=True)
df
```

```python
# Fillna Replace All Empty Values
df['born'].fillna(pd.Timestamp("2023-08-14"),inplace=True)
df
```
